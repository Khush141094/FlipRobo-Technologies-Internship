```python
driver=webdriver.Chrome(r"C:\Users\s0116731\Downloads\chromedriver_win32\chromedriver.exe")
driver.get('https://www.myntra.com/shoes?f=Color%3ABlack_36454f&rf=Price%3A7179.0_14119.0_7179.0%20TO%2014119.0')
# price_filter = driver.find_element_by_xpath("//ul[@class='price-list']/li/input[@value='7179.0 TO 14119.0']")
# price_filter.click()
# color_filter = driver.find_element_by_xpath("//li[@class='colour-listItem']/label[text()='Black']")
# color_filter.click()
```

```python
start_page=0
end_page=2
for page in range(start_page,end_page+1): 
    nxt_button=driver.find_element_by_xpath("//li[@class='pagination-next']")
    
    shoe_names=[]
    s_desc=[]
    short_desc=[]
    price=[]
    
    desc=driver.find_elements_by_xpath("//h3[@class='product-brand']")
    for i in desc:
        shoe_names.append(i.text)
        
    desc=driver.find_elements_by_xpath("//h4[@class='product-product']")
    for i in desc:
        s_desc.append(i.text)
   
#     for j in range(0,len(s_desc),2):
#         short_desc.append(s_desc[j])
    
    desc=driver.find_elements_by_xpath("//div[@class='product-price']")
    for i in desc:
        price.append(i.text)
    
    if nxt_button.text=='next':
            nxt_button.click()
            time.sleep(5)
    
    df=pd.DataFrame({'Brand': shoe_names,
                    'Short-description': s_desc,
                    'Price': price})
    print(df)
```
